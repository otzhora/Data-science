```python
from collections import defaultdict
import os
from pathlib import Path

import pandas as pd
```

```python
data_path = Path(os.getcwd()).parent
df = pd.read_csv(data_path/"data"/"titanic.csv")

df.head()
```

```python
male_count = df.loc[df.Sex == 'male'].PassengerId.count()
female_count = df.loc[df.Sex == 'female'].PassengerId.count()

print(male_count, female_count)
```

```python
woman_names = df.loc[df.Sex=='female'].Name.apply(lambda x: x.split())
```

```python
unique_words = defaultdict(int)

for s in woman_names:
    for word in s:
        if word[0] == "(" or word[0] == "\"":
            word = word[1:]
        if word[-1] == ',' or word[-1] == ")":
            word = word[:-1]
        unique_words[word] += 1

sorted(list(unique_words.items()), key = lambda x: x[0])
```
